```python
import pandas as pd
import numpy as np
import os
import re
from copy import copy
from time import sleep

import openai


import selenium
from bs4 import BeautifulSoup as bs

openai.api_key = input('input OPENAI api key here: ')
```

```python
categories = ['Production and Development',
 'Cultural Impact and Representation',
 'Reviews and Criticisms',
 'Marketing and Merchandising',
 'Audience Reception and Demographics',
 'Historical Context and Evolution of Barbie',
 'Interviews and Personal Insights',
 'Themes and Messages']
```

```python
def get_category(row):
    sleep(1)
    prompt = f'''"Fit the following article into one of the following catagories
{categories}
or
if you do not believe this to be about the barbie movie just return 'NA', but try and fit it into a catagory first


title = {row.title}
subtitle = {row.description}
content = {row.content}
'''

    response = openai.ChatCompletion.create(model="gpt-3.5-turbo", 
                                messages=[{'role':'user','content':prompt}],
                                temperature=0)
    print('next')
    return response.choices[0].message.content
```
