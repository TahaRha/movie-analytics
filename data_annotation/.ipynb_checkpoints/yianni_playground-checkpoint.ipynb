```python
for category, keywords in barbie_category_keywords.items():
    barbie_df[category] = False
    
    
for category, keywords in opp_category_keywords.items():
    op_df[category] = False

for category, keywords in mi_category_keywords.items():
    mi_df[category] = False
    
for index, row in barbie_df.iterrows():
    search_text = row['title'].lower() + " " + row['description'].lower()

    for category, keywords in barbie_category_keywords.items():
        if any(keyword in search_text for keyword in keywords):
            barbie_df.at[index, category] = True
            
for index, row in op_df.iterrows():
    search_text = row['title'].lower() + " " + row['description'].lower()

    for category, keywords in opp_category_keywords.items():
        if any(keyword in search_text for keyword in keywords):
            op_df.at[index, category] = True
            
for index, row in mi_df.iterrows():
    search_text = row['title'].lower() + " " + row['description'].lower()

    for category, keywords in mi_category_keywords.items():
        if any(keyword in search_text for keyword in keywords):
            mi_df.at[index, category] = True
```

```python
barbie_category_counts = barbie_df[list(barbie_category_keywords.keys())].sum()
opp_category_counts = op_df[list(opp_category_keywords.keys())].sum()
mi_category_counts = mi_df[list(mi_category_keywords.keys())].sum()

# Create a DataFrame from the counts
data = {'Barbie': barbie_category_counts, 'Oppenheimer': opp_category_counts, 'Mission Impossible':mi_category_counts}
df = pd.DataFrame(data)

# Plot the data
p = df.plot(kind='bar', figsize=(14, 8))

# Adding titles and labels
p.title('Category Counts Comparison between Barbie and Oppenheimer')
p.xlabel('Categories')
p.ylabel('Count')
p.xticks(rotation=45)
p.show()
```

```python

# Example text data preparation (as previously described)
barbie_text = barbie_df['title'] + " " + barbie_df['description']
opp_text = op_df['title'] + " " + op_df['description']
combined_text = barbie_text.tolist() + opp_text.tolist()

# Create and fit the TF-IDF Vectorizer
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(combined_text)

# Get feature names (words)
feature_names = vectorizer.get_feature_names()

# Convert the TF-IDF matrix for Barbie DataFrame into a readable format
barbie_tfidf_df = pd.DataFrame(tfidf_matrix[:len(barbie_text)].toarray(), columns=feature_names)

# Convert the TF-IDF matrix for Oppenheimer DataFrame into a readable format
opp_tfidf_df = pd.DataFrame(tfidf_matrix[len(barbie_text):].toarray(), columns=feature_names)

# Display TF-IDF scores for the first document in Barbie DataFrame
print("TF-IDF Scores for the first document in Barbie DataFrame:")
print(barbie_tfidf_df.iloc[0])

# # Display TF-IDF scores for the first document in Oppenheimer DataFrame
# print("\nTF-IDF Scores for the first document in Oppenheimer DataFrame:")
# print(opp_tfidf_df.iloc[0])
```
